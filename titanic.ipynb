```python
import pandas as pd
import numpy as np
import sklearn
import seaborn as sns

```

```python
df = pd.read_csv("train.csv")
df.head()
sns.countplot(df.Survived)
```

```python
df1 = df.drop('Name' , axis = 1)
```

```python
df2 = df1.dropna()
```

```python
df3 = df2.drop('Ticket' , axis = 1)
```

```python
df3['Sex'].replace({'female':1 , 'male':0} , inplace = True)
```

```python
df3
```

```python
df4 = df3.drop('Cabin' , axis = 1)
```

```python
df4 = df4.drop('Embarked' , axis = 1)
```

```python
df4 = df4.drop('PassengerId' , axis = 1)
```

```python
df4
```

```python
df5 = df4.drop('Survived',axis = 1)
x = df5
y = df4.Survived
x
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x,y, test_size = 0.5)
```

```python
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
prediction = clf.predict([[1,1,38,1,0,80]])
```
